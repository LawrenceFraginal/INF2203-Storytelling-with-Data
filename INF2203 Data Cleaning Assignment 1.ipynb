```python
# We'll start with the NYC vaccinations data with race. Main issues: irrelevant columns, bad column names, some outliers in the PERC1_PLUS column due to vaccination numbers surpassing estimated population count.
# Will eliminate superfluous columns, rename columns, and adjust the PERC_1PLUS so that numbers don't exceed 100%.

#Step 1: keep only relevant columns
vax2 = vax[['DATE', 'BOROUGH', 'AGE_GROUP', 'RACE_ETHNICITY', 'COUNT_1PLUS_CUMULATIVE', 'PERC_1PLUS']]
```

```python
# Step 2: adjust percentages in 1plus columns so that they don't surpass 100%
vax2['PERC_1PLUS'].mask(vax2['PERC_1PLUS'] > 100, 100, inplace=True)
```

```python
# review to ensure only the values in PERC_1PLUS > 100% have been adjusted to avoid interference with analysis
vax2
```

```python
#check for duplicates
vax2 = vax2.drop_duplicates()
vax2
# appears as though there were none
```

```python
# Save clean data set to CSV file
vax2.to_csv("C:/Users/FRAGINAL/Downloads/NYC2020vaxbyRaceandBoroclean.csv")
```

```python
# will do the same with the vaccinations data by ZIP code
vax_z2 = vax_z[['DATE', 'NEIGHBORHOOD_NAME', 'BOROUGH', 'MODZCTA', 'AGE_GROUP', 'COUNT_1PLUS_CUMULATIVE', 'PERC_1PLUS']]
```

```python
#finally, let's deal with the SQF data
# Step 1: Check for duplicates
sqf2 = sqf.drop_duplicates()
sqf2
#no duplicates found
```

```python
# Step 2: Get rid of irrelevant columns
sqf2 = sqf2[['STOP_ID','STOP_FRISK_DATE', 'STOP_FRISK_TIME', 'STOP_DURATION_MINUTES', 'OFFICER_IN_UNIFORM_FLAG', 'FRISKED_FLAG', 'SEARCHED_FLAG', 'SUSPECT_REPORTED_AGE','SUSPECT_SEX', 'SUSPECT_RACE_DESCRIPTION', 'STOP_LOCATION_X', 'STOP_LOCATION_Y', 'STOP_LOCATION_ZIP_CODE', 'STOP_LOCATION_BORO_NAME']]
```

```python
sqf2
```

```python
# Step 3: Drop missing values
# Note that there are several missing values in data set denoted by string '(null)'
# so we will have to drop those as well
sqf2.dropna(inplace=True)
```

```python
sqf2
```

```python
# to find nulls, we will check which columns are str types
sqf2.dtypes
```

```python
# will drop the '(null)' string from all those with the dtype object
# drop nulls
sqf2 = sqf2[sqf2['STOP_FRISK_DATE'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['STOP_FRISK_TIME'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['OFFICER_IN_UNIFORM_FLAG'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['FRISKED_FLAG'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['SEARCHED_FLAG'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['SUSPECT_REPORTED_AGE'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['SUSPECT_SEX'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['SUSPECT_RACE_DESCRIPTION'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['STOP_LOCATION_ZIP_CODE'].str.contains('(null)') == False]
sqf2 = sqf2[sqf2['STOP_LOCATION_BORO_NAME'].str.contains('(null)') == False]
```
